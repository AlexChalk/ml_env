```python
from itertools import combinations

population = np.random.exponential(size=10000)

fig, axis = plt.subplots()
axis.set_title("Population")
#ax.set_xticks([-10,-8,-6,-4,-2,0,2,4,6,8,10])
axis.hist(population, bins=300)
axis
```

C(n,r) = n! / (r! * (n-r)!)
e.g. C(10000,20) = 10000! / (20! * (10000-20)!) = 10000! / (20! * 9980!)
= ~2,745,184,200,000,000,000,000,000,000,000,000,000,000,000,000,000,000,000, or about 2.75 sexdecillion, or ~2.75 × 10^53.

```python
from itertools import combinations

def four_million_sample_means(population, sample_size):
    "A true sample distribution of means is too large to calculate in class"
    return np.mean(np.random.choice(population, size=(4000000, sample_size)), axis=1)
```

```python
sample_sizes = [1,2,4,10,30,60,120,200]

for sample_size in sample_sizes:
    fig, axis = plt.subplots()
    axis.set_title("Sample size: " + str(sample_size))
    axis.hist(four_million_sample_means(population, sample_size), bins=300)
```

```python
comp1 = np.random.standard_normal(5000)                                              
comp2 = 10 + 2 * np.random.standard_normal(5000)
population2 = np.concatenate([comp1, comp2])
                                                   
fig, axis = plt.subplots()
axis.set_title("Population")
axis.hist(population2, bins=300)
axis
```

```python
sample_sizes = [1,2,4,10,30,60,120,200]

for sample_size in sample_sizes:
    fig, axis = plt.subplots()
    axis.set_title("Sample size: " + str(sample_size))
    axis.hist(four_million_sample_means(population2, sample_size), bins=300)
```
