```python
import numpy as np, matplotlib.pyplot as plt
import math

def plot_function(f, min=0, max=5, color='r'):
    x = np.linspace(min,max, 100)[:,None]
    plt.plot(x, f(x), color)
```

```python
def func(x):
    return 5 - x**2 + 4*x
```

```python
np.log(0.001)
```

```python
plot_function(func, min=0,max=10)
```
